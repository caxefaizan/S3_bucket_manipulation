# mattplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[11,22,33,44]

#setup plot
fig,ax = plt.subplots(figsize=(10,10)) #figsize = width height
#plot data
ax.plot(x,y)
#customize plot
ax.set(title='simple plot',xlabel = 'x axis',ylabel='y-axis')

#save and show
fig.savefig("sample-plot.png")

```
